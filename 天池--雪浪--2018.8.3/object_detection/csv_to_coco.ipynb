```python
#将数据转换为coco格式
import json
import os
import sys
import cv2
```

```python
dataset={
    'licenses': [],
    'info': {},
    'categories': [],
    'images': [],
    'annotations': []
}
```

```python
classes=['bad']
for i,cls in enumerate(classes,1):
    dataset['categories'].append({
        'id': i,
        'name': cls,
        'supercategory': 'beverage'
    })
```

```python
def get_category_id(cls):
    for category in dataset['categories']:
        if category['name'] == cls:
            return category['id']
```

```python
dict1={}
#for n,i in enumerate(os.listdir("rpn_train/bad_aug/")):
for n,i in enumerate(os.listdir("val/val/bad")):
    dict1[i]=n
    print(n,i)
    dataset['images'].append({
        'coco_url':'',
        'date_captured': '',
        'file_name': i,
        'flickr_url': '',
        'id': n,
        'license': 0,
        'width': 2560,
        'height': 1920
    })
```

```python
import csv
with open("val.csv","r+") as csv_file:
    for j,line in enumerate(csv_file.readlines(),1):
        img_name,cls,x1,y1,x2,y2 = line[:-1].split(',')
        width = max(0, int(x2) - int(x1))
        height = max(0, int(y2) - int(y1))
        dataset['annotations'].append({
          'area': width * height,
          'bbox': [int(x1), int(y1), width, height],
          'category_id': get_category_id(cls),
          'id': j,
          'image_id': dict1[img_name],
          'iscrowd': 0,
          'segmentation': []
        })
```
